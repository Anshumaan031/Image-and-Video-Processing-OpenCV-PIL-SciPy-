# canny edge detection

```python
from matplotlib import pyplot as plt
import argparse
import cv2
```

```python
def plt_imshow(title, image):
   image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
   plt.imshow(image)
   plt.title(title)
   plt.grid(False)
   plt.show()
```

```python
#readind the image
image = cv2.imread('C:/Users/Administrator/Documents//ML/OpenCV/assignment/buildings/building1.jpg')
```

```python
plt_imshow('building',image)
```

```python
#applying gaussian blur
blur=cv2.GaussianBlur(image,(5,5),0)
plt_imshow('blurred',image)
```

```python
#converting the image to grayscale as canny edge detection only works with gray grayscale images.
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#applying canny edge detection to our building
edges=cv2.Canny(img_gray,100,200)
```

```python
plt_imshow('building',edges)
```
