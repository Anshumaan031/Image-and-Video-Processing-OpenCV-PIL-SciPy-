```python
# import the necessary packages
from matplotlib import pyplot as plt
import numpy as np
import argparse
import imutils
import cv2
```

```python
def plt_imshow(title,image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.grid(False)
    plt.show()
```

```python
image=cv2.imread("faces.jpg")
plt_imshow('faces',image)
```

```python
#cropping image
image_cropped=image[:250,:200]
plt_imshow('asian',image_cropped)
```

```python
image_flipped = cv2.flip(image, 1)
plt_imshow("Flipped Horizontally", image_flipped)
```

```python
image_flipped = cv2.flip(image, -1)
plt_imshow("Flipped Horizontally & Vertically", image_flipped)
```
